### A note on nwb terms and how they relate to our recording sessions:

#### `electrodes` are individual channels on a probe
- 1 `electrode` = 1 AP timeseries & 1 LFP timeseries
- typically, we acquire data from the 384 channels closest to the tip on a Neuropixels 1.0 probe
- each `electrode` will eventually have a `location` that reports the brain region it recorded from
- each `electrode` must belong to one `electrode group`

#### `electrode groups` represent probes inserted during a session
- we typically insert up to 6 probes in a session (probeA-probeF)
- each `electrode group` must be attached to a `device`    
- `devices` represent the identity of a physical probe across multiple sessions, e.g. its model & serial number 

```python
lfp = session.acquisition['LFP']['probeA']
display(lfp.electrodes[:].head())
display(lfp.electrodes[0].group[0])
display(lfp.electrodes[0].group[0].device)
```

when we plot the relative position of each electrode on the probe surface, we see 4 columns of channels running the length of the Neuropixels 1.0 probe:

```python
import matplotlib.pyplot as plt
import numpy as np

plt.cla()
fig, ax = plt.subplots(1,1)

for x in lfp.electrodes[:].rel_x.unique():
    rel_y = lfp.electrodes[:].query(f'rel_x == {x}').rel_y.values
    ax.scatter(np.full(len(rel_y), x), rel_y, marker='.')
ax.set_aspect(1/15)
ax.set_xticks(lfp.electrodes[:].rel_x.unique())
ax.set_xlabel('rel_x (microns)')
ax.set_ylabel('rel_y (microns): distance from tip')
secax = ax.secondary_yaxis('right')
secax.set_ylabel('channel number')
secax.set_yticks([0,  lfp.electrodes[:].rel_y.max()])
secax.set_yticklabels([lfp.electrodes[:].channel.min(), lfp.electrodes[:].channel.max()])
plt.suptitle('(x and y scales not equal)', fontsize=6)
plt.show()
```

as a result of this geometry, the data acquired from 4 electrodes in the same row will be largely similar, so we can generally take a slice of every 4th channel:

```python
import pynwb
import numpy.typing as npt


def plot_lfp(start_time: float, stop_time: float, lfp: pynwb.ecephys.LFP) -> plt.Figure:
    fig, ax = plt.subplots(1,1)
    electrodes_slice = slice(None, None, 4)
    data_to_millivolts = lfp.conversion*1e3
    im = ax.imshow(
        data_to_millivolts * lfp.data[round(start_time * lfp.rate): round(stop_time * lfp.rate), electrodes_slice].T,
        extent=[
            start_time,
            stop_time,
            lfp.electrodes[electrodes_slice].rel_y.min(),
            lfp.electrodes[electrodes_slice].rel_y.max(),
        ],
        # cmap='RdBu',
    )
    ax.set_aspect(1/lfp.rate)
    ax.set_xlabel('time (s)')
    ax.set_ylabel('distance from tip (microns)')
    plt.colorbar(im, label='mV')
    return fig

trial = session.trials[:].query('is_hit').iloc[0]
plot_lfp(
    start_time=trial.stim_start_time,
    stop_time=trial.response_window_stop_time,
    lfp=session.acquisition['LFP']['probeA'],
);
```
